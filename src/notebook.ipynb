```python
#imort libraries
import pandas as pd
import numpy as np

#load sample data set
df = pd.read_csv('../stackoverflow.csv')
df = df.sample(n = 1000, random_state=42).reset_index(drop=True)
df.drop('Unnamed: 0', axis=1, inplace=True)

print("Shape:", df.shape)  # Expected: (10000, 14)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values (%):\n", df.isnull().mean() * 100)
print("\nTarget Distribution:\n", df['Employed'].value_counts(normalize=True))
print("\nFirst Few Rows:\n", df.head())
```

```python
#impute missing values
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

num_cols = ['YearsCode', 'YearsCodePro', 'PreviousSalary']
cat_cols = ['Age', 'Accessibility', 'EdLevel', 'Employment', 'Gender', 'MentalHealth', 'MainBranch', 'Country']
multi_col = 'HaveWorkedWith'

num_imputer = SimpleImputer(strategy='median')
df[num_cols] = num_imputer.fit_transform(df[num_cols])

cat_imputer = SimpleImputer(strategy='most_frequent')
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

df[multi_col] = df[multi_col].fillna('None')
```

```python
#process multi-valus
df['SkillCount'] = df[multi_col].apply(lambda x: len(x.split(';')) if x != 'None' else 0)

all_tech = df[multi_col].str.split(';', expand=True).stack().value_counts()
top_tech = all_tech.head(20).index.tolist()
extra_tech = ["Rust", "Go", "Kotlin", "Swift", "Terraform", "Elasticsearch"] # Manually add rare-but-important techs
selected_tech = list(set(top_tech + extra_tech))

for tech in selected_tech:
    df[f'Tech_{tech}'] = df[multi_col].apply(lambda x: 1 if tech in x.split(';') else 0)

df.drop(multi_col, axis=1, inplace=True)
df.drop('ComputerSkills', axis=1, inplace=True)  # Drop if identical to ComputerSkills
```

```python
pd.set_option('display.max_columns', None)
print(df.head())
print("Shape after multi-val processing:", df.shape)
#(df['ComputerSkills'] == df['SkillCount']).all()  # True → identical, drop one
#(df['ComputerSkills'] - df['SkillCount']).describe()
```

```python
#encode and scale
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
scaler = StandardScaler()
df[num_cols + ['SkillCount']] = scaler.fit_transform(df[num_cols + ['SkillCount']])
```

```python
pd.set_option('display.max_columns', None)
print(df.head())
print("Final shape:", df.shape)
```

```python
# Cell 7: Save preprocessed data
df.to_csv('preprocessed_dataset.csv', index=False)
print("Preprocessed data saved.")
```
