```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('preprocessed_dataset.csv')
```

```python
sns.countplot(x='Employed', data=df)
plt.title('Employment Distribution')
plt.show()
```

```python
num_cols = ['YearsCode', 'YearsCodePro', 'PreviousSalary']
corr = df[num_cols + ['SkillCount', 'Employed']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.boxplot(x='Employed', y='YearsCode', data=df)
plt.title('Years of Code vs Employment')
plt.show()
```
